```python
root = "/content/drive/My Drive/Let-Hunt-Youtube/"
config_file_path = os.path.join(root, 'config.json')

config = json.load(open(config_file_path))

database_save_folder = config['folder']['database']
url_db_file_name = config['database']['url']

url_file_path = os.path.join(root, database_save_folder, url_db_file_name)
```

```python
url_df = pd.read_csv(url_file_path)
print(len(url_df))
utils.display_table(url_df)
```

```python
#@title Search by channel
channel_url = input('Channel URL: ')
new_df = utils.search_by_channel(channel_url=channel_url)
utils.display_table(new_df)
```

```python
#@title Search by keyword
keyword = input('Keyword: ')
new_df = utils.search_by_keyword(keyword=keyword)
utils.display_table(new_df)
```

```python
#@title Update database
url_df = utils.update_database(url_df, new_df)
utils.display_table(url_df)
```

```python
#@title Save database
assert url_df.video_url.nunique == len(url_df), print("duplicated data")
utils.save_database(url_df, url_file_path)
```
